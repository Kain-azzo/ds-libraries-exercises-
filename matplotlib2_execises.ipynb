```python
import matplotlib.pyplot as plt
```

```python
import math
x1 = range(100)
y1 = [(math.sqrt(n)) for n in x1]
plt.ylabel("Ys")
plt.xlabel("Xs")
plt.plot(x1,y1);
```

```python
x2 = range(100)
y2 = [(n**3) for n in x2]
plt.ylabel("Ys")
plt.xlabel("Xs")
plt.plot(x2,y2);
```

```python
x3 = range(100)
y3 = [(2**n) for n in x3]
plt.ylabel("Ys")
plt.xlabel("Xs")
plt.plot(x3,y3);
```

```python
x4 = range(100)
y4 = [1/(n + 1) for n in x4]
plt.ylabel("Ys")
plt.xlabel("Xs")
plt.plot(x4,y4);
```

```python
fig, ax = plt.subplots(2,2) #initialize at the top with the plt.subplots()

ax[0,0].plot(x1,y1) #call the ax variable and plot on it
ax[0,1].plot(x2,y2)
ax[1,0].plot(x3,y3)
ax[1,1].plot(x4,y4)

# ax[0,0].set_xlabel('babies') #calling ax specific plot for individual label

# fig.supxlabel('common x label') #calling the figure variable to get overall x-label
# fig.supylabel('common y label')
# fig.suptitle('how many grapes people eat by day')
plt>suptit
plt.tight_layout()
plt.show()
```
